```python
# 25f1001910@ds.study.iitm.ac.in
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider
from IPython.display import display, Markdown

# 25f1001910@ds.study.iitm.ac.in
# Cell 1: Dataset creation
data = pd.DataFrame({
    "x": np.arange(1, 101),
    "y": np.random.randn(100) * 10 + 50
})

display(Markdown("### Dataset Overview"))
display(data.head())

```

```python
# Cell 2: Slider & reactive outputs
def show_top_rows(top_n):
    # Dependent on Cell 1
    top_rows = data.head(top_n)

    # Display table
    display(Markdown(f"### Top {top_n} rows of the dataset"))
    display(top_rows)

    # Display mean
    mean_y = top_rows['y'].mean()
    display(Markdown(f"**Mean of y for top {top_n} rows:** {mean_y:.2f}"))

    # Plot
    plt.figure(figsize=(8,4))
    plt.plot(top_rows['x'], top_rows['y'], marker='o')
    plt.title(f"Top {top_n} rows: y vs x")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.show()

# Interactive slider
interact(show_top_rows, top_n=IntSlider(min=1, max=50, value=10, description="Top N"))

```
